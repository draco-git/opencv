```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 10.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
image1 =  cv2.imread('.\images\human_figure.jpg')
image = cv2.cvtColor(image1 , cv2.COLOR_BGR2RGB)
plt.imshow(image )
```

```python
gray_image = cv2.cvtColor(image , cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image , cmap = 'gray')
```

```python
smoothen_image = cv2.medianBlur(gray_image,7)
plt.imshow(smoothen_image , cmap = 'gray')
```

```python
edged_image = cv2.adaptiveThreshold(smoothen_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
plt.imshow(edged_image , cmap = 'gray')
```

```python
kernel = np.ones((5,5),np.uint8)
dilated_image = cv2.erode(edged_image , kernel , iterations = 1)
plt.imshow(dilated_image , cmap  ='gray')
```

```python
colorImage = cv2.bilateralFilter(image, 9, 600,600)
plt.imshow(colorImage , cmap = 'gray')
```

```python
cartoonImage = cv2.bitwise_and(colorImage, colorImage, mask=edged_image)
plt.imshow(cartoonImage , cmap = 'gray')
```

```python
cartoonImage = cv2.cvtColor(cartoonImage , cv2.COLOR_RGB2BGR)
image2 = cv2.resize(cartoonImage , (500,500))
cv2.imshow('cartoon',image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
im = cv2.imread('.\images\human_figure.jpg',cv2.IMREAD_COLOR)
imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,127,255,cv2.THRESH_BINARY)
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
for data in contours:
    print ("The contours have this data %r" %data)
cv2.drawContours(im,contours,-1,(0,255,0),3)
cv2.imshow('output',im)
while True:
    if cv2.waitKey(6) & 0xff == 27:
        cv2.destroyAllWindows()
        break
```
